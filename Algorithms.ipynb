```python
%matplotlib notebook
import matplotlib.pyplot as plt
import math
```

### Let's plot a graph in Python to check the complexity of different algorithms and it's behaviour : 
In the X axis we take the value of n and in the y axis we put the value of different algorithms such as O(logn), O($\sqrt{n}$), O(${n^2}$), O(${n^3}$), O(${2^n}$), O(n). This graph will show us the difference between these time functions.

```python
def plotting(k):
    if k>2:
        fig , ax = plt.subplots()
        n = [i for i in range(1,k)]
        logn = [math.log(i) for i in n]
        sqrtn = [math.sqrt(i) for i in n]
        sqrn = [i*i for i in n]
        cuben = [i*i*i for i in n]
        twopowern = [math.pow(2,i) for i in n]
        nlogn = [a*b for a,b in zip(n,logn)]
        

        ax.plot(n,n, label="O (n)")
        ax.plot(n,logn, label="O (logn)")
        ax.plot(n,sqrtn, label = "O (sqrt(n))")
        ax.plot(n,nlogn, label="O (nlogn)")
        ax.plot(n,sqrn, label="O (n^2)")
        ax.plot(n,cuben, label="O (n^3)")
        ax.plot(n,twopowern, label="O (2^n)")


        fig.suptitle("Representation of Time Functions")
        ax.set_xlabel("Value of n = %d"%(k-1))
        ax.set_ylabel("Value of graphs")
        ax.legend()
        fig.show()
    else:
        print("Give bigger input")
        
```

### This plotting funtion gives us a plot of these time functions and shows their graphs
plotting(10) will plot all the functions values up to 10 in a graph and plot it.


```python
plotting(10)
```

We can now run a loop to see various graphs to check the progress of the time funtions as the value of n increases.


```python
for i in range(3,20):
    plotting(i)
```
