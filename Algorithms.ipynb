```python
%matplotlib notebook
import matplotlib.pyplot as plt
import math
```

### Let's plot a graph in Python to check the complexity of different algorithms and it's behaviour : 
In the X axis we take the value of n and in the y axis we put the value of different algorithms such as O(logn), O($\sqrt{n}$), O(${n^2}$), O(${n^3}$), O(${2^n}$), O(n). This graph will show us the difference between these time functions.

```python
n = [i for i in range(1,7)]
logn = [math.log(i) for i in n]
sqrtn = [math.sqrt(i) for i in n]
sqrn = [i*i for i in n]
cuben = [i*i*i for i in n]
twopowern = [math.pow(2,i) for i in n]


plt.plot(n,logn, label="logn")
plt.plot(n,sqrtn, label = "Sqrt of n")
plt.plot(n,sqrn, label="Square of n")
plt.plot(n,cuben, label="Cube of n")
plt.plot(n,twopowern, label="2^n")


plt.title("Representation of Time Functions")
plt.xlabel("Value of n")
plt.ylabel("Value of graphs")
plt.legend()
plt.show()
```
